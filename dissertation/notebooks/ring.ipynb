```python
from sympy import *
```

```python
rc = Symbol("r_C", positive=True)
phic = Symbol("\\phi_C")
rcp = Symbol("\\dot{r_C}")
phicp = Symbol("\\dot{\\phi_C}")
dt = Symbol("\\delta t")
om = Symbol("\\omega")
```

```python
dx = (rc + rcp * dt) * cos(om + phic + phicp * dt) - rc * cos(om + phic)
dx
```

```python
limit(dx / dt, dt, 0)
```

```python
dy = (rc + rcp * dt) * sin(om + phic + phicp * dt) - rc * sin(om + phic)
dy
```

```python
limit(dy / dt, dt, 0)
```
