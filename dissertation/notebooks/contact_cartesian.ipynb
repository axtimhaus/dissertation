```python
from sympy import symbols, Symbol, sin, cos, asin, sqrt, limit, pi, plot, simplify
```

```python
r = Symbol("r", positive=True)
phi = Symbol("\\varphi")
omega = Symbol("\\omega")
domega = Symbol("\\delta \\omega")
alpha = Symbol("\\alpha", positive=True)
beta = Symbol("\\beta", positive=True)
dxP = Symbol("\\delta x_P")
dyP = Symbol("\\delta y_P")
xP = Symbol("x_P")
yP = Symbol("y_P")
ds = Symbol("\\delta s")
```

```python
xN = xP + r * cos(omega + phi)
xN
```

```python
yN = yP + r * sin(omega + phi)
yN
```

```python
dxN = ds * cos(omega + phi + (alpha + beta - pi))
dxN
```

```python
dyN = ds * sin(omega + phi + (alpha + beta - pi))
dyN
```

```python
dxR = r * cos(omega + phi + domega) - r * cos(omega + phi)
simplify(dxR)
```

```python
dxRlim = limit(dxR / domega, domega, 0)
dxRlim
```

```python
dyR = r * sin(omega + phi + domega) - r * sin(omega + phi)
simplify(dyR)
```

```python
dyRlim = limit(dyR / domega, domega, 0)
dyRlim
```

```python
dx = dxP + dxN + dxRlim * domega
dx
```

```python
dy = dyP + dyN + dyRlim * domega
dy
```
