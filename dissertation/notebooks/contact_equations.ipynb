```python
from sympy import symbols, Symbol, sin, cos, asin, sqrt, limit, pi, plot
```

```python
r = Symbol("r", positive=True)
ds, eta = symbols("ds eta")
```

```python
rp = sqrt(r**2 + ds**2 - 2 * r * ds * cos(eta))
rp
```

```python
dphi = asin(sin(eta)/rp * ds)
dphi
```

```python
limit(dphi/ds, ds, 0)
```

```python
simpl = dphi.replace(r, 1).replace(eta, pi/2)
simpl
```

```python
plot(simpl/ds, (ds, -0.1, 0.1))
```

```python
limit((rp - r)/ds, ds, 0)
```
