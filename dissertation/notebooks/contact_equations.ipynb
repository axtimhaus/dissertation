```python
from sympy import symbols, Symbol, sin, cos, asin, sqrt, limit, pi, plot, simplify
```

```python
r = Symbol("r", positive=True)
ds, eta = symbols("ds eta")
```

```python
rp = sqrt(r**2 + ds**2 - 2 * r * ds * cos(eta))
rp
```

```python
dphi = asin(sin(eta)/rp * ds)
dphi
```

```python
limit(dphi/ds, ds, 0)
```

```python
simpl = dphi.replace(r, 1).replace(eta, pi/2)
simpl
```

```python
plot(simpl/ds, (ds, -0.1, 0.1))
```

```python
limit((rp - r)/ds, ds, 0)
```

```python
r, dom, phi, phic, rc = symbols("r d\\omega phi phi_c r_c")
```

```python
ds = r / sin((pi - dom)/2) * sin(dom)
ds
```

```python
eta = -(phi - phic) + (pi - dom) / 2
eta
```
