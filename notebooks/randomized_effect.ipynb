```python
import pandas as pd
```

```python
# shrinkage 1e-6
nom = 1.189
df = pd.DataFrame(
    dict(
        mean=[1.156, 1.139, 1.185],
        std=[0.428, 0.364, 0.468],
    ),
    index=["circ", "oval", "shape"],
)
df["effect"] = df["mean"] - nom
df["cohen"] = df["effect"] / df["std"]
df
```
