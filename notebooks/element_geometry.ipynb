```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from sympy import *

plt.ioff()
```

```python
r = Symbol("r", positive=True)
ru = Symbol("r_u", positive=True)
rl = Symbol("r_l", positive=True)
au = Symbol("a_u", positive=True)
al = Symbol("a_l", positive=True)
alphau = Symbol("alpha_u", positive=True)
alphal = Symbol("alpha_l", positive=True)
phi = Symbol("phi", positive=True)
phiu = Symbol("phi_u", positive=True)
phil = Symbol("phi_l", positive=True)
```

```python
au_eq = Eq(au, sqrt(r**2 + ru**2 - 2 * r * ru * cos(phiu - phi)))
au_fn = lambdify([r, ru, phi, phiu], au_eq.rhs)
au_eq
```

```python
alphau_eq = Eq(alphau, acos((ru**2 - r**2 - au**2) / (-2 * r * au)))
alphau_fn = lambdify([r, ru, phi, phiu], alphau_eq.rhs.subs({au: au_eq.rhs}))
alphau_eq
```

```python
au_eq.rhs.evalf()
```

```python
@interact()
def plot_geometry(r_=1, ru_=1, rl_=1):
    fig, ax = plt.subplots(2, 1)

    angles = np.linspace(-0.5, 0.5, 20)

    ax[0].plot(angles, au_fn(r_, ru_, 0, angles))
    ax[1].plot(angles, alphau_fn(r_, ru_, 0, angles))

    return fig
```
