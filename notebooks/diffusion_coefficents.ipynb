```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temperatures = np.linspace(0, 2000 + 273, 300)
```

```python
def arrhenius(temp, pre, act):
    return pre * np.exp(-act / (8.314 * temp))
```

```python
fig, ax = plt.subplots()
```

```python
ax.plot(temperatures, arrhenius(temperatures, 1, 100e3))
```

```python
ax.plot(temperatures, arrhenius(temperatures, 1, 60e3))
```
