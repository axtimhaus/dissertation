```python
from sympy import *
```

```python
r = Symbol("r", positive=True)
o = Symbol("o", positive=True)
phi = Symbol("\\phi")
```

```python
a = sqrt(o) * r
b = r / sqrt(o)
```

```python
(a * b).simplify()
```

```python
(a / b).simplify()
```

```python
radius = a * b / sqrt((a * sin(phi)) ** 2 + (b * cos(phi)) ** 2)
```

```python
radius
```

```python
integrate(radius**2 / 2, (phi, 0, 2 * pi))
```

```python
integrate(r**2 / 2, (phi, 0, 2 * pi))
```

```python
import numpy as np
import scipy.integrate as spi

R = 1
O = 2
```

```python
spi.quad(lambda t: (r**2 / 2).evalf(subs={r: R, o: O, phi: t}), 0, 2 * np.pi)
```

```python
spi.quad(lambda t: (radius**2 / 2).evalf(subs={r: R, o: O, phi: t}), 0, 2 * np.pi)
```
